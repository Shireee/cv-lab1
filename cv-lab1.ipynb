```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image
image = cv2.imread('./cat.jpg', cv2.COLOR_BGR2RGB)
plt.imshow(image)

resolution = image.shape
height = image.shape[0]
width = image.shape[1]
channels = image.shape[2]
```

```python
R = [0 for x in range(256)]
G = [0 for x in range(256)]
B = [0 for x in range(256)]

def colorDistribution(image):
    # for RED
    for i in range(0, height):
        for j in range(0, width):
            value = image[i][j][0]
            R[value] += 1
    # for GREEN
    for i in range(0, height):
        for j in range(0, width):
            value = image[i][j][1]
            G[value] += 1
    # for BLUE
    for i in range(0, height):
        for j in range(0, width):
            value = image[i][j][2]
            B[value] += 1

colorDistribution(image)
```

```python
# Histograms for RGB channels
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.hist(R, bins=256, range=(0, 256), color='red', alpha=0.7)
plt.xlabel('Color Value')
plt.ylabel('Frequency')
plt.title('Red Channel Histogram')

plt.subplot(132)
plt.hist(G, bins=256, range=(0, 256), color='green', alpha=0.7)
plt.xlabel('Color Value')
plt.ylabel('Frequency')
plt.title('Green Channel Histogram')

plt.subplot(133)
plt.hist(B, bins=256, range=(0, 256), color='blue', alpha=0.7)
plt.xlabel('Color Value')
plt.ylabel('Frequency')
plt.title('Blue Channel Histogram')

plt.tight_layout()
plt.show()
```
